```python
#Reads .csv
#HOOK THIS IN TO FORMAT THE DATA YOU GRAB FROM THE PDF
CountyData = pd.read_csv('C:\\Users\\bague\\downloads\\county_reports_20200626.csv',delimiter=',',header=0, encoding='ascii',engine='python')

#forces data to numeric format and strips commas for large values. Does nothing special for smaller values
CountyData.replace(',','', regex=True, inplace=True)
CountyData['Cases'] = CountyData['Cases'].apply(pd.to_numeric,errors='coerce')
CountyData['Hospitalizations'] = CountyData['Hospitalizations'].apply(pd.to_numeric,errors='coerce')
CountyData['Deaths'] = CountyData['Deaths'].apply(pd.to_numeric,errors='coerce')

#creates a new dataframe with just Cases, Hospitals, Deaths
County = pd.DataFrame (columns = ['Cases','Hospitalizations','Deaths'])
County['Cases'] = CountyData['Cases']
County['Hospitalizations'] = CountyData['Hospitalizations']
County['Deaths'] = CountyData['Deaths']

#strips decimal displays if data decides it wan
pd.options.display.float_format = '{:,.0f}'.format
```
